```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import librosa
import librosa.display
import numpy as np

# Load dataset
df = pd.read_csv("/Users/kantharamoladayam/Documents/heart_disease/data/processed_features.csv")
print(df.head())

# Plot class distribution
plt.figure(figsize=(6, 4))
sns.countplot(x="label", data=df, palette="coolwarm")
plt.title("Class Distribution (Heart Disease vs No Heart Disease)")
plt.xlabel("Label")
plt.ylabel("Count")
plt.show()

# Visualizing feature correlations
plt.figure(figsize=(12, 8))
sns.heatmap(df.iloc[:, 1:-1].corr(), cmap="coolwarm", annot=False)
plt.title("Feature Correlation Heatmap")
plt.show()

# Example: Visualizing an audio file
audio_file = "data/records/13918_MV.wav"
y, sr = librosa.load(audio_file, sr=4000)
plt.figure(figsize=(10, 4))
librosa.display.waveshow(y, sr=sr)
plt.title("Waveform of a Sample Heart Sound")
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.show()

# Extracting sample MFCCs
mfccs = librosa.feature.mfcc(y=y, sr=sr, n_mfcc=13)
plt.figure(figsize=(10, 6))
librosa.display.specshow(mfccs, sr=sr, x_axis="time", cmap="coolwarm")
plt.colorbar(label="MFCC Coefficients")
plt.title("MFCCs of the Sample Heart Sound")
plt.show()
```
